```python
# Importing the cells geo dataset
cells_geo = pd.read_csv('/content/cells_geo.csv', sep=';', delimiter=None)
# Preview the first ten rows
cells_geo.head(5)

```

```python
# Importing the first Telcom dataset
# Using local path
telcom1 = pd.read_csv('/content/Telcom_dataset.csv')
# Preview the first ten rows
telcom1.head(5)
```

```python
# Importing the 2nd Telcom dataset
# Using local path
telcom2 = pd.read_csv('/content/Telcom_dataset2.csv')
# Preview the first ten rows
telcom2.head(5)
```

```python
# Importing the 3rd Telcom dataset
# Using local path
telcom3 = pd.read_csv('/content/Telcom_dataset3.csv')
# Preview the first ten rows
telcom3.head(5)
```

```python
# After a careful analysis of the telcom dataset, the columns are similar so they will be merged.
# Telcom 1 dataset has a typing error on 'PRODUTC' instead of 'PRODUCT'
# renaming column 'PRODUTC' in telcom 1 dataset
telcom1.columns = ['PRODUCT' if x=='PRODUTC' else x for x in telcom1.columns]
telcom1.head(5)
```

```python
# Merging the Telcom datasets
telcom = [telcom1,telcom2,telcom3]
tm_merge = pd.concat(telcom)
tm_merge.head(10)
```
